```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
img = cv2.imread('ucla.jpg', 0)
plt.imshow(img, cmap='gray')
```

**SVD provides the low rank approximation of input matrix(aka data) such that the approximation error is minimized. This fundamental matrix decomposition technique is ubiqitious among many machine learning applications.**


```python
U, S, V = np.linalg.svd(img, full_matrices=False)
U.shape, S.shape, np.diag(S).shape, V.shape 
```

```python
#define number of components
n_components = [768, 600, 400, 100, 85, 40]
plt.figure(figsize = (21, 10))

for i in range(len(n_components)):

    #new low rank matrix
    new_img = U[:, :n_components[i]] @ np.diag(S[:n_components[i]]) @ V[:n_components[i], :]    
    plt.subplot(2, 3, i+1)
    plt.title("n_components = {0}".format(n_components[i]))
    plt.imshow(new_img, cmap='gray')
```

Even the image with n_components = 85 is fairly distinguishable.
